#SECTION 04: Preprocessing

```python
df = df[df['Age'] <= 90]
print("Shape after outlier removal:", df.shape)
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single'
})
print(df['Marital_Status'].unique())
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
print("Missing values remaining:", df['Income'].isnull().sum())
```

```python
education_order = {
    'Basic': 0,
    'Graduation': 1,
    '2n Cycle': 2,
    'Master': 3,
    'PhD': 4
}
df['Education'] = df['Education'].map(education_order)
print(df['Education'].unique())
```

```python
df = pd.get_dummies(df, columns=['Marital_Status'], drop_first=False)
print(df.shape)
```

```python
features = ['Income', 'Age', 'Total_Spent', 'Total_Purchases',
            'Children', 'Recency', 'Education']
df_cluster = df[features]
print(df_cluster.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_cluster)
print("Shape:", df_scaled.shape)
```

#SECTION 5: ML

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)
print("PCA Shape:", df_pca.shape)
print("Variance:", pca.explained_variance_ratio_)
```

```python
from sklearn.cluster import KMeans
inertia = []
K_range = range(1, 11)
for k in K_range:
   kmeans = KMeans(n_clusters=k, random_state=42)
   kmeans.fit(df_scaled)
   inertia.append(kmeans.inertia_)
plt.plot(K_range, inertia, marker='o')
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df_cluster['Cluster'] = kmeans.fit_predict(df_scaled)
print(df_cluster['Cluster'].value_counts())
```

###Now we will apply the silhouette method to see that if k=3 is actually a good choice to proceed with. We got the score of 0.228 rounded off, which means the clusters are weak.

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(df_scaled, df_cluster['Cluster'])
print("Silhouette Score:", score)
```

```python
cluster_summary = df_cluster.groupby('Cluster').mean()
print(cluster_summary)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df_pca[:,0], df_pca[:,1], c=df_cluster['Cluster'], cmap='viridis')
plt.title('Customer Segments (K-Means) - PCA Visualization')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
plt.figure(figsize=(12,6))
linked = linkage(df_scaled, method='ward')
dendrogram(linked)
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
df_cluster['HC_Cluster'] = hc.fit_predict(df_scaled)
print(df_cluster['HC_Cluster'].value_counts())
```

```python
score_hc = silhouette_score(df_scaled, df_cluster['HC_Cluster'])
print("Hierarchical Silhouette Score:", score_hc)
```

###With the refined feature set, K-Means (0.229) outperformed Hierarchical Clustering (0.182) in terms of cluster separation, making K-Means the preferred algorithm for this segmentation task.
